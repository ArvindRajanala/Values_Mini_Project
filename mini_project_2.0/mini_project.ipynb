# Importing the Text Files

```python
import os
import re


  
# Source Folder Path
path = r'C:\Users\2069718\OneDrive - Cognizant\Documents\mini_project_2.0\sources'

# Loading all values in a list using patten matching
pattern = re.compile(r'[A-Za-z-]+')
values_sample = r'C:\Users\2069718\OneDrive - Cognizant\Documents\mini_project_2.0\sources\values_sample.txt'
all_values = []
for i, line in enumerate(open(values_sample)):
    for match in re.finditer(pattern, line):
        all_values.append(match.group().lower())


# Change the directory
os.chdir(path)
  
# Funtionn to Read text File
values = []
def read_text_file(file_path):
    # using pattern matcher and all_values list to collect values from all text files
    interim=[]
    count=1
    with open(file_path, 'r') as f:
        for i, line in enumerate(f.read().split()):
            for match in re.finditer(pattern, line):
                word = match.group().lower()
                #checking if the word is in the all_values list, avoiding duplicates and only taking 10 values from a file
                if word in all_values and word not in interim and count <=10:
                    interim.append(word)
                    count+=1
        values.extend(interim)


# iterate through all file
for file in os.listdir():
    # Check whether file is in text format and starts with F
    if file.endswith(".txt") and file.startswith("F"):
        file_path = f"{path}\{file}"
  
        # call read text file function
        read_text_file(file_path)

# for debugging 
print(len(values))
for value in values:
    print(value,'\n')
        
```

# Importing Excel file using Pandas

```python
#Using pandas to import excel file

import pandas as pd
file21 = r'C:\Users\2069718\OneDrive - Cognizant\Documents\mini_project_2.0\sources\file 21.xlsx'
read_file = pd.read_excel(file21,header=None)

#using Stack method to shift 12 columns into a single column
read_file = read_file.stack().reset_index()
read_file = read_file.drop(['level_0','level_1'],axis=1)

#converting dataframe object to list
excel_file = list(read_file[0])
excel_file
```

# Joining the Text file and Excel file values

```python
values.extend(map(lambda x:x.lower(),excel_file))
len(values)
```

# Calculating Counts of each Value

```python
value_and_count = {}
for value in values:
    if value not in value_and_count:
        value_and_count[value] = 1
    else:
        value_and_count[value]+=1

#sorting the dictionary based on descending count order
value_and_count = dict(sorted(value_and_count.items(), key=lambda item: item[1],reverse=True))
print(value_and_count)
```

# Saving the result as text file

```python
#storinh the dictionary in a dataframe
df=pd.DataFrame(value_and_count.items(), columns=['VALUE', 'COUNT'])

#storing only the top ten values based on count
df=df.head(10)

#saving the top ten values in a csv file
target_path = r'C:\Users\2069718\OneDrive - Cognizant\Documents\mini_project_2.0\target\top-ten.txt'
df.to_csv(target_path, index=None, sep=',', mode='w')
```
